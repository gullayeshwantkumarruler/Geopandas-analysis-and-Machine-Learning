####`Task 1.2.4 Plot (scatter or line plot) the percentage change in renewable electricity output from 2000 to 2010, for  the country having the highest population in 2010. In this question, you must consider the renewable electricity output for each year between 2000 and 2010. Very briefly interpret the generated plot.`

```python

# Filter the population data to consider only the year 2010
pop_data_2010 = pop1_data[['Country Name', 'Country Code', '2010']]

# Find the country with the highest population in 2010
highest_population_country = pop_data_2010.loc[pop_data_2010['2010'].idxmax()]

# Extract the country code
country_code = highest_population_country['Country Code']

# Filter the electricity data for the specific country
elc_data_country = elc_data[elc_data['Country Code'] == country_code]

# Extract renewable electricity output for the years 2000 and 2010
renewable_output_2000 = elc_data_country[2000].values[0]
renewable_output_2010 = elc_data_country[2010].values[0]

# Calculate the percentage change
pct_change = ((renewable_output_2010 - renewable_output_2000) / renewable_output_2000) * 100

# Plotting
plt.figure(figsize=(10, 6))
plt.plot([2000, 2010], [renewable_output_2000, renewable_output_2010], marker='o')
plt.title(f"Percentage Change in Renewable Electricity Output (2000-2010) for {highest_population_country['Country Name']}")
plt.xlabel('Year')
plt.ylabel('Renewable Electricity Output (%)')
plt.xticks([2000, 2010])
plt.grid(True)
plt.annotate(f'{pct_change:.2f}%', xy=(2005, (renewable_output_2000 + renewable_output_2010) / 2),
             xytext=(2000, (renewable_output_2000 + renewable_output_2010) / 2), ha='center', fontsize=12)
plt.show()
```

The above plot illustrates the percentage change in renewable electricity output for a specific country between the years 2000 and 2010. The diagonal line that starts from the bottom left corner and goes towards the top right corner of the plot represents this change. The x-axis of the plot represents the years 2000 and 2010, while the y-axis represents the renewable electricity output percentage.

The plot's purpose is to visually display the trend in renewable electricity output for the chosen country over the specified timeframe. The line's slope and direction indicate whether the country's renewable electricity output has increased or decreased during this period.

The title of the plot indicates the country for which the analysis is conducted. The x-axis labels show the years being analyzed (2000 and 2010), and the y-axis labels indicate the renewable electricity output percentage. The annotation on the plot provides additional context by showing the calculated percentage change in renewable electricity output between the two years.

The diagonal line's direction and slope reveal whether the country's renewable electricity output has experienced growth or decline over the decade. If the line moves upwards from left to right, it indicates a positive trend of increased renewable electricity output. Conversely, if the line moves downwards, it signifies a reduction in renewable electricity output.

Thus, the plot provides a concise and visually informative representation of the percentage change in renewable electricity output for a specific country between 2000 and 2010, allowing for easy interpretation of the country's renewable energy trends over the specified period.

#### `Task 1.2.5 Present a scatter plot between the mean population of each country and mean renewable electricity output considering from year 2000 until 2010. Very briefly interpret the generated plot, particularly look for any correlation (if present) among the plotted variables. In this question, you have to consider each year between 2000 and 2010 to find the mean population and mean renewable electricity output`

```python
# Filter the population and electricity data to consider years from 2000 to 2010
years = list(range(2000, 2011))
pop_data_filtered = pop_data[['Country Name', 'Country Code'] + years]
elc_data_filtered = elc_data[['Country Code'] + years]

# Merge the filtered population and electricity data on 'Country Code'
merged_data = pd.merge(pop_data_filtered, elc_data_filtered, on='Country Code')

# Rename columns to remove suffixes '_x' and '_y'
merged_data.rename(columns=lambda x: x.replace('_x', ''), inplace=True)
merged_data.rename(columns=lambda x: x.replace('_y', ''), inplace=True)
columns = merged_data.columns
years = [int(col) if col.isdigit() else col for col in columns]
merged_data.columns = years

# Calculate the mean population and mean renewable electricity output for each country
merged_data['Mean Population'] = merged_data[years].mean(axis=1)
merged_data['Mean Renewable Electricity Output'] = merged_data[years].mean(axis=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Mean Population'], merged_data['Mean Renewable Electricity Output'], alpha=0.7)
plt.title("Scatter Plot: Mean Population vs. Mean Renewable Electricity Output (2000-2010)")
plt.xlabel("Mean Population")
plt.ylabel("Mean Renewable Electricity Output")
plt.grid(True)
plt.show()
```

The generated scatter plot illustrates the relationship between the mean population and the mean renewable electricity output for various countries over the years 2000 to 2010. Each point on the scatter plot represents a country, with its x-coordinate corresponding to the country's mean population and its y-coordinate corresponding to the mean renewable electricity output.

The scatter plot reveals interesting insights regarding the correlation between these two variables. Upon visual inspection, it appears that there is a slight positive correlation between mean population and mean renewable electricity output. This observation suggests that countries with larger populations tend to have higher mean renewable electricity outputs.

The distribution of points in the scatter plot also indicates a broad distribution of mean renewable electricity output values for countries with varying population sizes. This suggests that other factors beyond population size could be influencing the mean renewable electricity output, contributing to the overall variability in the data.

```python

def get_axis(location):
    try:
        loc = geolocator.geocode(location, language='en')
        if loc:
            return loc.latitude, loc.longitude
        else:
            return None
    except:
        return None
```

```python
tweet_data['axis_xy']=tweet_data['userLocation'].apply(get_axis)
```

```python
import nltk
import re
import string
from nltk.corpus import stopwords
stop_words = stopwords.words("english")
def text_preproc(x):
    x = re.sub("@\S+", " ", x)
    x = ' '.join([word for word in x.split(' ') if word not in stop_words])
    x = re.sub("https*\S+", " ", x)
    x = re.sub("#\S+", " ", x)
    x = re.sub("\'\w+", '', x)
    x = re.sub('[%s]' % re.escape(string.punctuation), ' ', x)
    x = re.sub(r'\w*\d+\w*', '', x)
    x = re.sub('\s{2,}', " ", x)
    return x

tweet_data['cleaned_tweets'] = tweet_data['tweetContent'].apply(text_preproc)
```

#### Task 2.3 Polarity analysis

Calculate the polarity values of all the tweets. For a given geographical location, if you have more than

```python
from textblob import TextBlob
def polarity(text):
   return  TextBlob(text).sentiment.polarity
tweet_data['Polarity'] = tweet_data['cleaned_tweets'].apply(polarity)
```

```python
import pandas as pd
import plotly.express as px


# Group the data by location and calculate the average polarity for each location
avg_polarity = tweet_data.groupby(['userLocation'])['Polarity'].mean().reset_index()

# Create the chloropleth map using Plotly
fig = px.choropleth(avg_polarity, locations='userLocation', locationmode='country names',
                    color='Polarity', range_color=(-1, 1),
                    title='Average Polarity by Location',
                    color_continuous_scale=px.colors.diverging.RdYlBu_r)

fig.show()
```

The above output visualizes the average polarity of tweets based on their respective user locations through a choropleth map. The map is divided into different regions corresponding to the user locations. The color intensity on the map represents the average polarity of the tweets originating from each location.

As per the color scale used (RdYlBu_r), areas with bluish tones indicate more negative sentiments, while areas with reddish tones represent more positive sentiments. The color gradient between these extremes captures the varying degrees of polarity.

By observing this map, we can glean insights into the overall sentiment trends across different geographic locations. The map can help identify regions that are more inclined towards positive or negative sentiments. This visualization is valuable for understanding the distribution of sentiment across various locations, potentially unveiling patterns related to local events, cultural contexts, or regional sentiments.
